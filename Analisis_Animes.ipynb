```python
#Generamos una funcion para reparar los valores nulos en el dataframe:

def reparar_nulos(df, col):
    if df[col].dtype == 'int64' or df[col].dtype == 'float64': #si los valores de la columna son de tipo numerico, los valores nulos de esa columna seran reemplazados por la media.
        df[col] = df[col].fillna(df[col].mean())
    else:
        df[col] = df[col].fillna(df[col].mode()[0]) #si los valores de la columna NO son de tipo numerico, los valores nulos de esa columna seran reemplazados por la moda.
    return df 

for col in df.columns:
    reparar_nulos(df, col)
```

# Analizamos los datos resolviendo algunas preguntas:

### 1- En funcion de lo que dura su episodio, son mas populares los animes cortos o largos? Considerar unicamente series de tv.

```python
#Seteamos para imprimir todas las filas del dataframe

pd.set_option('display.max_rows', None)
```

```python
#Imprimimos la cantidad de valores que tenemos en la columna "type":

df['type'].value_counts()
```

```python
# Creamos el dataframe dftv donde nos quedamos solo series de tv en la columna "type":

dftv = df[df['type'] == 'TV']  
```

```python
#imprimos el nuevo dftv para comprobar que se haya llevado a cabo el filtrado:

dftv.head(5)
```

```python
#Ahora que tenemos unicamente las series de TV, tenemos que comparar la duracion. Para eso, nos fijamos los valores unicos de la columna "duration" para ver que hay que corregir:

dftv['duration'].unique()
```

```python
#eliminamos el texto despues del numero entero para quedarnos solo con el numero.

dftv['duration'] = dftv['duration'].str.replace(' min. per ep.', '') 
dftv['duration'] = dftv['duration'].str.replace(' min per ep.', '')
dftv['duration'] = dftv['duration'].str.replace(' min.', '')
dftv['duration'] = dftv['duration'].str.replace('  min per ep', '')
dftv['duration'] = dftv['duration'].str.replace('per ep', '')
dftv['duration'] = dftv['duration'].str.replace(' min', '')
dftv['duration'] = dftv['duration'].str.replace(' min.', '')
```

```python
#Filtramos las filas donde la columna 'duration' no contiene la cadena 'sec. per ep':

dftv = dftv[~dftv['duration'].str.contains('sec')]
dftv = dftv[~dftv['duration'].str.contains('sec.')]
dftv = dftv[~dftv['duration'].str.contains('sec per ep')]
dftv = dftv[~dftv['duration'].str.contains('sec. per ep')]
```

```python
#Filtramos las filas donde la columna 'duration' no contiene la cadena 'Unknown':

dftv = dftv[~dftv['duration'].str.contains('Unknown')]
```

```python
#Imprimimos la columna "duration" para ver los valores unicos luego de los cambios anteriores:

dftv['duration'].unique()
```

```python
#convertimos los valores de la columna "duration" a valores numericos:

dftv['duration'] = pd.to_numeric(dftv['duration'])
```

```python
#imprimimos la columna "duration" para comprobar los cambios y que sean todos valores numericos:

dftv['duration']
```

```python
#Como queremos comprobar si los episodios son cortos o largos, con "describe" imprimimos las estadísticas de los valores de la columna "duration":

dftv['duration'].describe()
```

```python
#vemos los rangos intercuartilicos
sns.boxplot(dftv['duration'])
```

```python
#Como resultado de las estadisticas vistas anteriormente, como el 50% de las series de TV son de 24min, creamos dos grupos, las series que duran mas de 24 minutos y las que duran menos:

dftv['duration'] = np.where(dftv['duration'] > 24, 'Largo', 'Corto') #si la duracion es de mas de 24min, la serie se considera "largo", sino es "corto"

dftv['duration'].value_counts() #contamos la cantidad de series de TV
```

```python
#Ahora comprobamos con nuestra columna score si los animes cortos o largos son mas populares usando pandas 

dftv.groupby('duration')['score'].mean().round(2)
```

```python
#Cmprobamos con nuestra columna popularity si los animes cortos o largos son mas populares usando pandas 

dftv.groupby('duration')['popularity'].mean().round(2)
```

Conclusion: Tanto las series de TV cortas como las largas tienen un score similar, pero teniendo en cuenta la popularidad, las series de TV largas son mas populares. Para una mayor precision en la conclusion, sera necesario profundizar en detalle estadisticamente.
